```python
# ---- OLD (article) imports — DO NOT USE, kept here for reference ----
# from keras.datasets import imdb
# from keras.models import Sequential
# from keras.layers import Dense, Flatten
# from keras.layers.convolutional import Conv1D, MaxPooling1D   # REMOVED in Keras 3
# from keras.layers.embeddings import Embedding                  # REMOVED in Keras 3
# from keras.preprocessing import sequence                       # MOVED in Keras 3

# ---- UPDATED imports that work with TensorFlow 2.x / Keras 3 ----
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv1D, MaxPooling1D, Embedding
from tensorflow.keras.utils import pad_sequences  # sequence.pad_sequences moved here

print(f"TensorFlow version: {tf.__version__}")
print(f"Keras version: {tf.keras.__version__}")
```

```python
# Our dictionary will contain only the top 7000 words appearing most frequently
top_words = 7000

# Now we split our data-set into training and test data
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=top_words)

# Looking at the nature of training data
print(X_train[0])
print(y_train[0])
print('Shape of training data: ')
print(X_train.shape)
print(y_train.shape)
print('Shape of test data: ')
print(X_test.shape)
print(y_test.shape)
```

```python
# Padding the data samples to a maximum review length in words
max_words = 450
X_train = pad_sequences(X_train, maxlen=max_words)
X_test  = pad_sequences(X_test,  maxlen=max_words)
```

```python
# Building the CNN Model
model = Sequential()      # initializing the Sequential nature for CNN model

# Embedding layer: maps top 7000 words -> 32-dimensional vectors, max review = 450 words
model.add(Embedding(top_words, 32, input_length=max_words))
model.add(Conv1D(32, 3, padding='same', activation='relu'))
model.add(MaxPooling1D())
model.add(Flatten())
model.add(Dense(250, activation='relu'))
model.add(Dense(1, activation='sigmoid'))  # binary classification -> 1 output node

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
# Fitting the data onto model
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=2, batch_size=128, verbose=2)

# Getting score metrics from our model
scores = model.evaluate(X_test, y_test, verbose=0)

# Displays the accuracy of correct sentiment prediction over test data
print("Accuracy: %.2f%%" % (scores[1]*100))
```
